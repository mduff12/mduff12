```python
# https://amueller.github.io/word_cloud/auto_examples/masked.html

d = path.dirname('./data/total_cleaned.csv') if "__file__" in locals() else os.getcwd()

text = open(path.join(d, './data/total_cleaned.csv')).read()

mask = np.array(Image.open(path.join(d,"./images/qanon.jpg")))

stopwords = set(STOPWORDS)
stopwords.add('now')
stopwords.add('people')
stopwords.add('S')
stopwords.add('think')

wc = WordCloud(background_color="white", max_words=2000, mask=mask,
               stopwords=stopwords, contour_width=3, contour_color='steelblue')

wc.generate(text)


wc.to_file(path.join(d, "./images/newq.png"))

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
