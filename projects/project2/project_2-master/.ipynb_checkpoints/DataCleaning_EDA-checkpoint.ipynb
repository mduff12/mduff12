## Prepare

**Import Data**
- using Pandas
- are there values that pandas should avoiding making null? 

        For example:
        Pandas imports NA as null, when in reality some of the NA's mean something in the data. To override the auotmatic creation of nulls from NA,:*

        < df = pd.read_csv('datasets/train.csv', keep_default_na=False, na_values=['']) >

        makes sure that "NA" is NOT converted to np.NaN, and empty strings WILL be converted to np.NaN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split, cross_val_score
```

## Testing correlation between features to see if they are collinear:
 - I suspect that some of the numerical variables have relationships to one another. I want to make sure they do not throw off the model if they are related to eachother. 

```python
house[['Garage Cars','Garage Area']].corr()
```

#### Garage area has a slightly higher correlation to saleprice(.66 v .67) , and garage area and garage cars are very highly correlated (.90). I will drop *Garage Cars* from the model. 

```python
house[['Total Bsmt SF','1st Flr SF']].corr()
```

#### Total basement sf has a slightly higher correlation to saleprice (.67 v .68), and basement sf and 1st floor sf are highly correlated (.80). I will drop 1st Flr SF from the model. 

```python
house[['Year Built','Year Remod/Add']].corr()
```

```python
house[['Year Built','Garage Yr Blt']].corr()
```

#### Year Built has a slightly higher correlation to saleprice (.58 v .54), and year build and garage yr blt are strongly correlated (.80). I will drop garage yr blt from the model. 

Columns To Drop:

- 1st Flr SF
- Garage Cars
- Garage Yr Blt
